## 1. ¿Cuantos trámites tienen como requisito el C.I.?

```python
# Considerar que en este data frame por ser desglosado, un trámite puede constar de varias filas pero los identificadores 'id' se mantienen únicos por trámite
unique_ids = df[df['requisito.nombre']=='Cédula de Identidad']['id'].unique()
print(f"Trámites que requieren Cédula de identidad: {len(unique_ids)}\n\n{unique_ids}")
```

```python
import re

regex_str = r".*(C\.I\.)|(carnet de identidad)|(Cédula de Identidad).*"
pattern_ci = re.compile(regex_str, re.IGNORECASE)

_df = df.copy()
_df['tieneCi'] = _df.apply(lambda x: bool(pattern_ci.match(x['requisito.nombre'])), axis=1)
print(_df[_df['tieneCi']==True].shape)
_df[_df['tieneCi']==True]
```

```python
unique_ids = _df[_df['tieneCi']==True]['id'].unique()
print(f"Trámites que requieren Cédula de identidad: {len(unique_ids)}\n\n{unique_ids}")
```

```python
df_tramites = pd.read_csv('../tramites.csv')
print(df_tramites.shape)
print(df_tramites.columns)
```

```python
import json
import ast

df['enlaces'] = df['enlaces'].map(lambda x: json.dumps(ast.literal_eval(x), ensure_ascii=False))
df['ubicaciones'] = df['ubicaciones'].map(lambda x: json.dumps(ast.literal_eval(x), ensure_ascii=False))
df.sample(2)
```

```python

df_tramites['modalidades'].map(lambda x: json.dumps(ast.literal_eval(x), ensure_ascii=False))[0]
```

```python
def buscar_regex_en_requisitos(raw_modalidad, pattern, campo_requisito='nombre'):
    """Busca el patron dado en todos los requisitos del tramite. 
    Returns: Devuelve la cantidad de veces que se encuentra el patron en los procedimientos
    """
    count_matches = 0
    modalidades = ast.literal_eval(raw_modalidad)
    for modalidad in modalidades:
        for mod_publico in modalidad['publico']:
            for requisito in mod_publico['requisitos']:
                matched = pattern.match(requisito[campo_requisito])
                # print(requisito[campo_requisito], matched)
                if matched is not None:
                    count_matches += 1
    return count_matches


import re

regex_str = r".*(C\.I\.)|(carnet de identidad)|(Cédula de Identidad).*"
pattern_ci = re.compile(regex_str, re.IGNORECASE)

buscar_regex_en_requisitos(df_tramites.iloc[0]['modalidades'], pattern_ci, 'nombre')
```

```python
regex_str = r".*(C\.I\.)|(carnet de identidad)|(Cédula de Identidad).*"
pattern_ci = re.compile(regex_str, re.IGNORECASE)

df_tramites['ciEncontrados'] = df_tramites.apply(lambda x: buscar_regex_en_requisitos(x['modalidades'], pattern_ci, 'nombre'), axis=1)
df_tramites[['id', 'nombre', 'entidad.sigla', 'modalidades', 'ciEncontrados']]
```

```python
unique_ids = df_tramites[df_tramites['ciEncontrados']>0]['id'].unique()
print(f'Número de trámites que tienen al C.I. como requisito: {len(unique_ids)}: \n{unique_ids}')
```

```python
regex_str = r".*(fotocopia)|(copia de).*"
pattern_ci = re.compile(regex_str, re.IGNORECASE)

_df['ciFotocopiasEncontrados'] = df_tramites[df_tramites['ciEncontrados']>0] \
    .apply(lambda x: buscar_regex_en_requisitos(x['modalidades'], pattern_ci, 'nombre'), axis=1)
#df_tramites[['id', 'nombre', 'entidad.sigla', 'modalidades', 'ciEncontrados']]
_df[_df['ciFotocopiasEncontrados']>0]
```

## 3. ¿Cuantos trámites tienen como requisito la ciudadanía digital?


```python
regex_str = r".*(Ciudadania digital)|(Ciudadanía Digital).*"
pattern_cd = re.compile(regex_str, re.IGNORECASE)

_df = df_tramites.copy()
_df['tieneCiudadania'] = _df.apply(lambda x: buscar_regex_en_requisitos(x['modalidades'], pattern_cd, 'nombre'), axis=1)
print(_df[_df['tieneCiudadania']==True].shape)
_df[_df['tieneCiudadania']==True]
```

```python
unique_ids_ciudadania = _df[_df['tieneCiudadania']==True]['id'].unique()
print(f"Trámites que requieren Ciudadanía digital: {len(unique_ids_ciudadania)}\n\n{unique_ids_ciudadania}")
```
