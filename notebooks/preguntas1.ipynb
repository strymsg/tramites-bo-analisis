```python
def buscar_regex_en_requisitos(raw_modalidad, pattern, campo_requisito='nombre'):
    """Busca el patron dado en todos los requisitos del tramite. 
    Returns: Devuelve la cantidad de veces que se encuentra el patron en los procedimientos
    """
    count_matches = 0
    modalidades = ast.literal_eval(raw_modalidad)
    for modalidad in modalidades:
        for mod_publico in modalidad['publico']:
            for requisito in mod_publico['requisitos']:
                matched = pattern.match(requisito[campo_requisito])
                # print(requisito[campo_requisito], matched)
                if matched is not None:
                    count_matches += 1
    return count_matches


import re

regex_str = r".*(C\.I\.)|(carnet de identidad)|(Cédula de Identidad).*"
pattern_ci = re.compile(regex_str, re.IGNORECASE)

buscar_regex_en_requisitos(df_tramites.iloc[0]['modalidades'], pattern_ci, 'nombre')
```

```python
regex_str = r".*(C\.I\.)|(carnet de identidad)|(Cédula de Identidad).*"
pattern_ci = re.compile(regex_str, re.IGNORECASE)

df_tramites['ciEncontrados'] = df_tramites.apply(lambda x: buscar_regex_en_requisitos(x['modalidades'], pattern_ci, 'nombre'), axis=1)
df_tramites[['id', 'nombre', 'entidad.sigla', 'modalidades', 'ciEncontrados']]
```

```python
unique_ids = df_tramites[df_tramites['ciEncontrados']>0]['id'].unique()
print(f'Número de trámites que tienen al C.I. como requisito: {len(unique_ids)}: \n{unique_ids}')
```

## 2. ¿Cuantos trámites tienen como requisito la fotocopia de C.I.?


```python
df['requisito.nombre'].unique()
```

Probando en el dataset tramites.csv que parece más completo a pesar de estar menos desglosado.

```python
regex_str = r".*(fotocopia)|(copia).*|(copiado)"
pattern_fotocopia_ci = re.compile(regex_str, re.IGNORECASE)

# fotocopia de CI en el nombre del requisito
s1 = df[df['tieneCi']==True].apply(lambda x: probar_regex(x['requisito.nombre'], pattern_fotocopia_ci), axis=1)
unique_ids1 = s1[s1==True].index
print(f'Número de trámites que tienen la fotocopia de C.I. como nombre de requisito: {len(unique_ids1)}: \n{unique_ids1}')

# fotocopia de CI en el comentario del requisito
s2 = df[df['tieneCi']==True].apply(lambda x: probar_regex(x['requisito.comentario'], pattern_fotocopia_ci), axis=1)
unique_ids2 = s2[s2==True].index
print(f'Número de trámites que tienen la fotocopia de C.I. como comentario de requisito: {len(unique_ids2)}: \n{unique_ids2}')

# Identificadores de trámites donde la fotocopia de C.I. es mencionada en el nombre o comentario
unique_ids = list(set(list(unique_ids1) + list(unique_ids2)))
print(f'Número de trámites que requieren la fotocopia de C.I.: {len(unique_ids)}: \n{unique_ids}')

dict_respuestas['tieneFotocopiaCi'] = {
    'cantidad': len(unique_ids),
    'id_tramites': unique_ids
}
df.loc[unique_ids][['id', 'nombre', 'entidad.sigla', 'entidad.nombre', 'requisito.nombre', 'requisito.comentario']]
```

## 3. ¿Cuantos trámites tienen como requisito la ciudadanía digital?


```python
regex_str = r".*(Ciudadania digital)|(Ciudadanía Digital).*"
pattern_cd = re.compile(regex_str, re.IGNORECASE)

s = df.apply(lambda x: probar_regex(x['requisito.nombre'], pattern_cd), axis=1)
unique_ids = s[s==True].index
print(f"Trámites que requieren Ciudadanía digital: {len(unique_ids)}\n\n{unique_ids}")

dict_respuestas['ciudadaniaDigital'] = {
    'cantidad': len(unique_ids),
    'id_tramites': unique_ids
}
df.loc[unique_ids][['id', 'nombre', 'entidad.sigla', 'entidad.nombre', 'requisito.nombre']]
```

```python
for key in dict_respuestas.keys():
    print(f'{key}: {dict_respuestas[key]['cantidad']}')
```
