# Trámites del estado Boliviano (gob.bo) - Desglosando o expandiendo el dataset

El dataset original tramites.json está muy anidado y la idea de este notebook es desglosarlo o expandirlo para facilitar el análisis. Para esto se ha visto que cada trámite puede tener varios:

- Públicos objetivo
- Procedimientos divididas por modalidades
- Duración de trámites
- Costos
- Requisitos




```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

df = pd.read_csv('../tramites.csv')
df
```

## Dicionarios, listas, tuplas -- guradar como JSON

En este dataset muchas columnas como listas u objetos json, al ser leídas se guardan como diccionarios python representados en cadenas. Para estandarizar el dataset para que sea usable en otros lenguages más fácilmente todas las columnas con listas, tuplas o diccionarios se guardarán como cadenas JSON.

```python
import json
import ast

df['enlaces'] = df['enlaces'].map(lambda x: json.dumps(ast.literal_eval(x), ensure_ascii=False))
df['ubicaciones'] = df['ubicaciones'].map(lambda x: json.dumps(ast.literal_eval(x), ensure_ascii=False))
df.sample(2)
```

## Desglosando modalidades

```python
import json
import ast

ast.literal_eval(df.iloc[0]['modalidades'])
```

```python
def desglosar_segun_tipo_modalidad(df):
    new_df = pd.DataFrame(columns=df.columns.to_list() + ['modalidad.tipo', 'modalidad'])
    print(new_df.columns)
    _df = df.copy()
    for i in _df.index.to_list():
        data = _df.iloc[i].to_dict()
        #print(data)
        modalidades = ast.literal_eval(data['modalidades'])
        for modalidad in modalidades:
            data['modalidad.tipo'] = modalidad['tipo']
            data['modalidad'] = modalidad
            #print(f"Agregando id: {data['id']} -> tipo: {data['modalidad.tipo']}")
            new_df.loc[len(new_df)] = data
    del new_df['modalidades']
    return new_df

new_df = desglosar_segun_tipo_modalidad(df)
print(new_df.shape)
print(f'¿Misma cantidad de trámites?: {len(new_df['id'].unique().tolist()) == len(df['id'].unique().tolist())}')
new_df.sample(3)
```

```python
# Estan todos los id de trámites del dataframe original?
set(new_df['id'].unique().tolist()).issubset(df['id'].unique().tolist())
```

## Dividiendo en procedimientos (una fila para cada público objetivo)

Al parecer hay también diferentes públicos objetivo por cada trámite. Para facilitar el análisis conviene dividir cada trámite también por público objetivo, donde cada uno de estos tiene su propio **procedimiento**. Entonces un trámite puede tener varias modalidades y cada modalidad varios procedimientos.

```python
def desglosar_por_procedimientos(df):
    _new_df = pd.DataFrame(
        columns=df.columns.to_list() + 
        [
         'tramite.url', 'tramite.urlSeguimiento',
         'tramite.tieneCosto', 'tramite.duracion', 'tramite.unidadMedidaDuracion', 'tramite.procedimiento',
         'tramite.publicoObjetivo', 'tramite.requisitoNoParametrizado', 'tramite.estado',
         'tramite.emisionFactura','tramite.formaCobro','tramite.horarios','tramite.depositoBancario',
         'tramite.costos', 'tramite.requisitos'
         ])
    print(_new_df.columns)
    _df = df.copy()
    for i in _df.index.to_list():
        data = _df.iloc[i]['modalidad']
        for procedimiento in data['publico']:
            proc = _df.iloc[i].copy()
            proc['tramite.url'] = data['url']
            proc['tramite.urlSeguimiento'] = data['urlSeguimiento']
            proc['tramite.tieneCosto'] = 1 if procedimiento['tieneCosto'] == True else 0
            proc['tramite.duracion'] = procedimiento['duracion']
            proc['tramite.unidadMedidaDuracion'] = procedimiento['unidadMedidaDuracion']
            proc['tramite.procedimiento'] = procedimiento['procedimiento']
            proc['tramite.publicoObjetivo'] = procedimiento['nombrePublicoObjetivo']
            proc['tramite.requisitoNoParametrizado'] = procedimiento['requisitoNoParametrizado']
            proc['tramite.estado'] = procedimiento['estadoPublicoObjetivo']
            proc['tramite.emisionFactura'] = procedimiento['emisionFactura']
            proc['tramite.formaCobro'] = procedimiento['formaCobro']
            proc['tramite.horarios'] = procedimiento['horario']
            proc['tramite.depositoBancario'] = procedimiento['depositoBancario']
            proc['tramite.costos'] = procedimiento['costos']
            proc['tramite.requisitos'] = procedimiento['requisitos']
            del proc['modalidad']
            #print(f"Agregando id: {_df.iloc[i]['id']} --> {proc['tramite.publicoObjetivo']}")
            _new_df.loc[len(_new_df)] = proc
    del _new_df['modalidad']
    return _new_df

new_df = desglosar_por_procedimientos(new_df)
print(new_df.shape)
print(f'¿Misma cantidad de trámites?: {len(new_df['id'].unique().tolist()) == len(df['id'].unique().tolist())}')
new_df.sample(3)
```

```python
def obtener_duracion(num_duracion, unidad_medida, unidad_destino='min'):
    minutos = 0
    try:
        num_duracion = int(num_duracion)
    except Exception as E:
        return np.nan
    
    if unidad_medida == 'Día(s)':
        minutos = num_duracion*1440
    elif unidad_medida == 'Horas(s)':
        minutos = num_duracion*60
    elif unidad_medida == 'Mes(es)':
        minutos = num_duracion*43200
    elif unidad_medida == 'Año(s)':
        minutos = num_duracion*15768000
    elif unidad_medida == 'Semana(s)':
        minutos = num_duracion*10080

    if unidad_destino == 'min':
        return int(minutos)
    elif unidad_destino == 'hora':
        return int(minutos/60)
    elif unidad_destino == 'dia':
        return int(minutos/1440)

new_df['tramite.duracionMin'] = new_df.apply(
    lambda x: obtener_duracion(x['tramite.duracion'], x['tramite.unidadMedidaDuracion'], 'min'), axis=1)
new_df['tramite.duracionHora'] = new_df.apply(
    lambda x: obtener_duracion(x['tramite.duracion'], x['tramite.unidadMedidaDuracion'], 'hora'), axis=1)
new_df['tramite.duracionDia'] = new_df.apply(
    lambda x: obtener_duracion(x['tramite.duracion'], x['tramite.unidadMedidaDuracion'], 'dia'), axis=1)

print(f'¿Misma cantidad de trámites?: {len(new_df['id'].unique().tolist()) == len(df['id'].unique().tolist())}')
new_df.sample(5)[['tramite.duracion', 'tramite.duracionMin', 'tramite.duracionHora', 'tramite.duracionDia',
                       'tramite.unidadMedidaDuracion', 'tramite.unidadMedidaDuracion',
                      'modalidad.tipo', 'estado']]
```

## Desglosando costos de tramites

```python
s = new_df.groupby('tramite.tieneCosto').count()['id']
s.index = ['Sin costo', 'Tiene costo']
s
```

```python
def desglosar_segun_costos(df):
    _new_df = pd.DataFrame(columns=df.columns.to_list() + [
        'costo.conceptoPago', 'costo.monto', 'costo.moneda'])
    print(_new_df.columns)
    _df = df.copy()
    for i in _df.index.to_list():
        data = _df.iloc[i].copy().to_dict()
        #print(data)
        costos = data['tramite.costos']
        if len(costos) == 0:
            data['costo.conceptoPago'] = ''
            data['costo.monto'] = np.nan
            data['costo.moneda'] = ''
            _new_df.loc[len(_new_df)] = data
        else:
            for costo in costos:
                data['costo.conceptoPago'] = costo['conceptoPago']
                data['costo.monto'] = float(costo['costo'])
                data['costo.moneda'] = costo['moneda']
                #print('>>>>', data)
                #print(f"Agregando id: {data['id']} -> {data['costo.monto']}: {data['costo.moneda']}")
                _new_df.loc[len(_new_df)] = data
    del _new_df['tramite.costos']
    return _new_df

new_df = desglosar_segun_costos(new_df)
print(new_df.shape)
print(f'¿Misma cantidad de trámites?: {len(new_df['id'].unique().tolist()) == len(df['id'].unique().tolist())}')
print(len(new_df['id'].unique().tolist()), len(df['id'].unique().tolist()))
new_df.sample(3)
```

## Requisitos y procedimientos

Reto: Averiguar cúantos trámites y procedimientos requieren el Carnet de Identidad (C.I) como requisito.

```python
def desglosar_segun_costos(df):
    _new_df = pd.DataFrame(columns=df.columns.to_list() + [
        'requisito.comentario', 'requisito.nombre', 'requisito.entidad', 'requisito.entidadActiva'])
    print(_new_df.columns)
    _df = df.copy()
    for i in _df.index.to_list():
        data = _df.iloc[i].to_dict()
        #print(data)
        requisitos = data['tramite.requisitos']
        if len(requisitos) == 0:
            data['requisito.comentario'] = ''
            data['requisito.nombre'] = ''
            data['requisito.entidad'] = ''
            data['requisito.entidadActiva'] = 0
            _new_df.loc[len(_new_df)] = data
            continue
        for requisito in requisitos:
            data['requisito.comentario'] = requisito['comentario']
            data['requisito.nombre'] = requisito['nombre']
            data['requisito.entidad'] = requisito['entidad']
            data['requisito.entidadActiva'] = 1 if requisito['entidadActiva'] == True else 0
            #print('>>>>', data)
            #print(f"Agregando id: {data['id']} -> {data['requisito.nombre']}")
            _new_df.loc[len(_new_df)] = data
    del _new_df['tramite.requisitos']
    return _new_df

new_df = desglosar_segun_costos(new_df)
print(new_df.shape)
print(f'¿Misma cantidad de trámites?: {len(new_df['id'].unique().tolist()) == len(df['id'].unique().tolist())}')
new_df.sample(3)
```

## Guardando como csv

Solo quedaría pendiente guardar los campos que son diccionarios como cadenas JSON.


```python
df_desglosado = pd.DataFrame(columns=new_df.columns.to_list())

for column in new_df.columns:
    print(column, new_df[column].dtype)
    if new_df[column].dtype != 'object':
        df_desglosado[column] = new_df[column]
    else:
        # comprobando si la columna tiene objetos que pueden representarse como json
        contains_dict = new_df[column].map(lambda x: isinstance(x, dict) | isinstance(x, list) | isinstance(x, tuple))
        #print(f'{column}:\n {True in contains_dict.unique()}')
        if True in contains_dict.unique(): # existe al menos un valor que es diccionario
            print(f'Existe dict o array: {contains_dict.unique()}')
            try:
                df_desglosado[column] = new_df[column].map(lambda x: json.dumps(x, ensure_ascii=False))
            except:
                print(f'No se pudo convertir diccionario a json: {new_df[column]}')
                df_desglosado[column] = new_df[column]
        else:
            #print(f'No existe dict: {contains_dict.unique()}')
            df_desglosado[column] = new_df[column]

print(df_desglosado.shape, df_desglosado.columns)
df_desglosado.sample(3)
```

```python
df_desglosado.to_csv('../tramites_desglosados.csv', index=False)
```
