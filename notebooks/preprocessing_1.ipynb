# Preprocesamiento datos tramites.jsonl

```python
import json
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)

# Obteniendo data frame pandas para explorar
lines = []
with open(r'../tramites.jsonl') as f:
    lines = f.read().splitlines()

line_dicts = [json.loads(line) for line in lines]
df = pd.DataFrame(line_dicts)

df
```

```python
# columnas pendientes:
pendientes = ['palabrasClave', 'entidad', 'categorias', 'eventosVida',
              'herramientas', 'documentos',
             ]

new_df = df[[col for col in df.columns if col not in pendientes]].copy()
new_df.sample(3)
```

## Desglosando columnas

### entidad

```python
# desglosando columnas de la entidad

new_rows = {
    'entidad.nombre': [],
    'entidad.sigla': [],
    'entidad.slug': [],
    'entidad.logo': [],
    'entidad.descripcion': [],
    'entidad.urlSitioWeb': [],
    'entidad.telefonoContacto': [],
    'entidad.correoElectronicoContacto': [],
}
for i in range(0, df.shape[0]):
    entidad_json = df.iloc[i]['entidad']
    #entidad_json = json.loads(df.iloc[i]['entidad'])
    new_rows['entidad.nombre'].append(entidad_json['nombre'])
    new_rows['entidad.sigla'].append(entidad_json['sigla'])
    new_rows['entidad.slug'].append(entidad_json['slug'])
    new_rows['entidad.logo'].append(entidad_json['logo'])
    new_rows['entidad.descripcion'].append(entidad_json['descripcion'])
    new_rows['entidad.urlSitioWeb'].append(entidad_json['urlSitioWeb'])
    new_rows['entidad.telefonoContacto'].append(entidad_json['telefonoContacto'])
    new_rows['entidad.correoElectronicoContacto'].append(entidad_json['correoElectronicoContacto'])

for k, v in new_rows.items():
    new_df[k] = pd.Series(v).values

new_df.sample(3)
```

### esPresencial y esVirtual


```python
rows = []
for i in range(0, df.shape[0]):
    palabras_clave = df.iloc[i]['palabrasClave']
    if isinstance(palabras_clave, list):
        rows.append(','.join(palabras_clave))
    else:
        rows.append('')

new_df['palabrasClave'] = pd.Series(rows).values
new_df.sample(5)['palabrasClave']
```

### categorias

```python
rows = []
for i in range(0, df.shape[0]):
    categorias = df.iloc[i]['categorias']
    categorias_list = [categoria['nombre'] for categoria in categorias]
    if len(categorias_list) > 0:
        rows.append(','.join(categorias_list))
    else:
        # TODO: ver como normalizar valores vacios, NULL o cadena vacia?
        rows.append('')
new_df['categorias'] = rows
new_df.sample(5)['categorias']
```

### eventosVida

```python
def separar_lista_por_caracter(df, col, separator=','):
    rows = []
    for i in range(0, df.shape[0]):
        words = df.iloc[i][col]
        if isinstance(words, list):
            rows.append(separator.join(words))
        else:
            rows.append('')
    return rows
```

```python
rows = separar_lista_por_caracter(df, 'eventosVida')
new_df['eventosVida'] = pd.Series(rows).values
new_df.sample(5)['eventosVida']
```

### documentos

```python
rows = separar_lista_por_caracter(df, 'documentos')
new_df['documentos'] = pd.Series(rows).values
new_df.sample(5)['documentos']
```

```python
new_df['enlaces'] = df['enlaces']
```

### herramientas

```python
rows = separar_lista_por_caracter(df, 'herramientas')
new_df['herramientas'] = pd.Series(rows).values
new_df.sample(5)['herramientas']
```

### esPresencial

```python
new_df['esPresencial'] = new_df['esPresencial'].apply(lambda value: 1 if value is True else 0)
new_df['esVirtual'] = new_df['esVirtual'].apply(lambda value: 1 if value is True else 0)
new_df.sample(5)[['id', 'nombre', 'esPresencial', 'esVirtual']]
```
